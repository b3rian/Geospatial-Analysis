```python
import rasterio as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = st.open(r"C:\Users\ADMIN\Downloads\RGB.byte.tif")
```

```python
from rasterio.plot import show
from rasterio import open as rio_open
import rasterio
```

```python
# Stack bands into a 3D array
red = dataset.read(1)  # Read Red band
green = dataset.read(2)  # Read Green band
blue = dataset.read(3)
rgb_image = np.stack([red, green, blue], axis=-1)

# Normalize pixel values (0-255)
rgb_image = (rgb_image / np.max(rgb_image)) * 255
rgb_image = rgb_image.astype(np.uint8)

plt.figure(figsize=(10, 10))
plt.imshow(rgb_image)
plt.title("RGB Image from Raster")
plt.axis("off")
plt.show()
```

```python
red = dataset.read(1)
green = dataset.read(2)
blue = dataset.read(3)

# Plot individual bands
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(red, cmap="Reds")
axes[0].set_title("Red Channel")

axes[1].imshow(green, cmap="Greens")
axes[1].set_title("Green Channel")

axes[2].imshow(blue, cmap="Blues")
axes[2].set_title("Blue Channel")

plt.tight_layout()
plt.show()
```

```python
# Compute NDBI (approximate)
ndbi = (red.astype(float) - green.astype(float)) / (red + green + 1e-5)

# Plot NDBI
fig, ax = plt.subplots(figsize=(8, 8))
ndbi_img = ax.imshow(ndbi, cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(ndbi_img, ax=ax, label="NDBI")
ax.set_title("Built-up Index (NDBI)")
plt.show()
```
